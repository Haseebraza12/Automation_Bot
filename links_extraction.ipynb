```python
import csv
import difflib
from urllib.parse import urlparse

def classify_links(input_csv, output_csv, similarity_threshold=0.5):
    """
    Classify similar links from an input CSV and save them to an output CSV.
    
    :param input_csv: Path to the input CSV file containing links.
    :param output_csv: Path to the output CSV file for similar links.
    :param similarity_threshold: Threshold for considering links as similar (0.0 to 1.0).
    """
    try:
        # Read links from the input CSV
        with open(input_csv, mode='r', encoding='utf-8') as file:
            reader = csv.reader(file)
            next(reader)  # Skip header
            links = [row[0] for row in reader if row]

        # Group similar links
        grouped_links = []
        visited = set()
        
        for i, link1 in enumerate(links):
            if link1 in visited:
                continue

            group = [link1]
            visited.add(link1)

            for link2 in links[i+1:]:
                if link2 not in visited:
                    # Use difflib to calculate similarity
                    similarity = difflib.SequenceMatcher(None, link1, link2).ratio()
                    if similarity >= similarity_threshold:
                        group.append(link2)
                        visited.add(link2)

            grouped_links.append(group)

        # Write grouped links to the output CSV
        with open(output_csv, mode='w', newline='', encoding='utf-8') as file:
            writer = csv.writer(file)
            writer.writerow(["Group", "Links"])  # Header
            for idx, group in enumerate(grouped_links, start=1):
                writer.writerow([f"Group {idx}", ", ".join(group)])

        print(f"Similar links have been saved to {output_csv}")

    except Exception as e:
        print(f"Error processing links: {e}")

def main():
    input_csv = "extracted.csv"
    output_csv = "similar_links_1.csv"
    classify_links(input_csv, output_csv)

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
from fuzzywuzzy import fuzz
from collections import defaultdict

# Load CSV files
csv1 = pd.read_csv("similar_links_1.csv")
csv2 = pd.read_csv("similar_links_2.csv")
csv3 = pd.read_csv("similar_links_3.csv")  # Replace with actual name if different

# Combine all links into a single list
def extract_links(df):
    links = []
    for link_list in df['Links']:
        links.extend(link_list.split(", "))  # Split comma-separated URLs
    return links

all_links = set(extract_links(csv1) + extract_links(csv2) + extract_links(csv3))

# Group similar URLs
def group_similar_links(links, threshold=85):
    groups = defaultdict(list)
    visited = set()
    
    for link in links:
        if link in visited:
            continue
        
        group = [link]
        visited.add(link)
        
        for other_link in links:
            if other_link not in visited:
                similarity = fuzz.ratio(link, other_link)
                if similarity >= threshold:  # If similarity exceeds the threshold
                    group.append(other_link)
                    visited.add(other_link)
        
        groups[f"Group {len(groups) + 1}"] = group
    
    return groups

# Create grouped links
grouped_links = group_similar_links(all_links)

# Convert to a DataFrame and save to a CSV
grouped_data = [{"Group": group, "Links": ", ".join(links)} for group, links in grouped_links.items()]
grouped_df = pd.DataFrame(grouped_data)

# Save results
grouped_df.to_csv("grouped_links.csv", index=False)
print("Grouped links saved to 'grouped_links.csv'")
```
